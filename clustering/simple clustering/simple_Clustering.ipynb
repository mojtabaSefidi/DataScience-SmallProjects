```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
sn.set()
from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv("Countries-exercise.csv")
raw_data.head()
```

```python
data = raw_data.copy()

```

```python
plt.scatter(data["Longitude"],data["Latitude"])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

```python
x = data.iloc[:,1:]
x.head()
```

```python
kmeans = KMeans(5)
```

```python
kmeans.fit(x)
```

```python
identified_cluster = kmeans.fit_predict(x)
identified_cluster
```

```python
clustered_data = data.copy()
clustered_data["Cluster"] = identified_cluster
clustered_data
```

```python
plt.scatter(data['Longitude'], data['Latitude'],c=clustered_data['Cluster'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```
