```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
observation = 1000
x1 = np.random.uniform(-10,10,(observation,1))
x2 = np.random.uniform(-10,10,(observation,1))
inputs = np.column_stack((x1,x2))
inputs.shape
```

```python
noise = np.random.uniform(-1.5,1.5,(observation,1))
target = 5*x1 + 7*x2 - 10 + noise
target.shape
```

```python
target = target.reshape(observation,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x1,x2,target)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Targets')
ax.view_init(azim=200)
plt.show()
target = target.reshape(observation,1)
```

```python
w_range = 0.1
weights = np.random.uniform(-w_range,w_range,(2,1))
bias = np.random.uniform(-w_range,w_range,1)
print(weights)
print(bias)
learning_rate = 0.04
```

```python
for i in range(100):
    output = np.dot(inputs,weights)+bias
    delta = output-target
    loss = np.sum(delta**2)/2/observation
    print(loss)
    delta_scaled = delta/observation
    weights = weights - learning_rate * np.dot(inputs.T,delta_scaled)
    bias = bias - learning_rate * np.sum(delta_scaled)
```

```python
print(weights)
print(bias)
```

```python
plt.plot(output,target)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
