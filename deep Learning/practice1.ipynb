```python
import numpy as np
import tensorflow as tf
```

```python
observation = 1500
x1 = np.random.uniform(-10,10,(observation,1))
x2 = np.random.uniform(-10,10,(observation,1))
inputs = np.column_stack((x1,x2))
noise = np.random.uniform(-1,1,(observation,1))
targets = -2 * x1 + 6 * x2 + 13 + noise
np.savez('data',inputs=inputs,targets=targets)
```

```python
training_data = np.load("data.npz")
inputSize = 2
outputSize = 1
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(outputSize,
                                                   kernel_initializer=tf.random_uniform_initializer(-0.1,0.1),
                                                   bias_initializer=tf.random_uniform_initializer(-0.1,0.1)
                                                   )
                             ])
optimizer = tf.keras.optimizers.SGD(0.02)
model.compile(optimizer=optimizer,loss="mean_squared_error")
model.fit(training_data["inputs"],training_data["targets"],epochs=200,verbose=2)
```
