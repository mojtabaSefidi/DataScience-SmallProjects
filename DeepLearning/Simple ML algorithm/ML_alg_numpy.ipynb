```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
observation = 1000

x1 = np.random.uniform(low=-10,high=10,size=(observation,1))
x2 = np.random.uniform(low=-10,high=10,size=(observation,1))
inputs = np.column_stack((x1,x2))
inputs.shape
```

```python
noise = np.random.uniform(low=-1,high=1,size=(observation,1))
targets = 2*x1 - 3*x2 + 5 + noise
targets.shape
```

```python
targets = targets.reshape(observation,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x1, x2, targets)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('Targets')
ax.view_init(azim=100)
plt.show()
targets = targets.reshape(observation,1)
```

```python
init_range = 0.1
weights = np.random.uniform(low=-init_range,high=init_range,size=(2,1))
bias = x1 = np.random.uniform(low=-init_range,high=init_range,size=1)
print(weights)
print(bias)
```

```python
learning_rate = 0.02
```

```python
for i in range(100):
    outputs = np.dot(inputs,weights) + bias
    deltas = outputs - targets
    loss = np.sum(deltas ** 2) / 2 /observation
    print(loss)
    deltas_scaled = deltas / observation
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    bias = bias - learning_rate * np.sum(deltas_scaled)

```

```python
print(weights,bias)
```

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
